```python
import pandas as pd
import numpy as np
```

```python
book_df = pd.read_csv('/content/book.csv',encoding='latin-1')
book_df
```

```python
book_df.head()
```

```python
#Renaming The coloum index names 
book_df=book_df.rename({'User.ID':'userid','Unnamed: 0':'index','Book.Title':'bookTitle','Book.Rating':'bookRating'},axis=1)
book_df
```

```python
#number of unique users in the dataset
len(book_df.userid.unique())
```

```python
book_df[book_df.duplicated()] # Checking for duplicated values
```

```python
len(book_df.bookTitle.unique()) #Getting booktitle unique values
```

```python
book_df[book_df.duplicated(subset=["userid","bookTitle"])]
```

```python
#Removing Duplicated Data
book_df=book_df.drop_duplicates(subset=["userid","bookTitle"])
book_df
```

```python
user_book_df = book_df.pivot(index='userid',
                                 columns='bookTitle',
                                 values='bookRating').reset_index(drop=True)
```

```python
user_book_df
```

```python
#Impute those NaNs with 0 values
user_book_df.fillna(0, inplace=True)
```

```python
user_book_df
```

```python
#Calculating Cosine Similarity between Users
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine, correlation
```

```python
reader = 1 - pairwise_distances( user_book_df.values,metric='cosine')
reader
```

```python
#Store the results in a dataframe
reader_df = pd.DataFrame(reader)
reader_df
```

```python
#Set the index and column names to user ids 
reader_df.index = book_df.userid.unique()
reader_df.columns = book_df.userid.unique()
```
